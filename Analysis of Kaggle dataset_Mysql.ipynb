```python
import pandas as pd
import pandasql as ps
```

```python
df = pd.read_csv("Kaggle_Datasets.csv")
df.head()
```

```python
# Finding count of files types

ans = ps.sqldf("""
SELECT

f.types_of_files,
COUNT(f.num_of_files) AS nf
FROM df as f
GROUP BY f.types_of_files

""")

ans
```

```python
# Finding the titles having more than 75000 downloads

ans = ps.sqldf("""
SELECT
f.title,
f.downloads
FROM df as f
WHERE f.downloads>75000

""")
ans
```

```python
# Titles with 1 lakh views and its usability

ans = ps.sqldf("""

SELECT

f.title,
f.usability,
f.downloads
FROM df AS f
WHERE f.views>1000000

""")

ans
```

```python
# 10 records of Titles and medals with the keyword 'Business'

ans = ps.sqldf("""

SELECT f.title,
f.medal
FROM df AS f
WHERE f.keywords = 'business' LIMIT 10


""")
ans
```

```python
# Selecting titles and usability having vote counts 100 , 200 & 400

ans = ps.sqldf("""

SELECT f.title,
f.vote_counts,
f.usability
FROM df AS f
WHERE f.vote_counts IN (100,200,400);


""")
ans
```

```python
# Filtering Top 10 'title' and 'downloads' having range of 10000 to 15000 downloads

ans = ps.sqldf("""

SELECT f.title,
f.downloads
FROM df AS f
WHERE f.downloads BETWEEN 10000 AND 15000
ORDER BY f.downloads DESC limit 10;

""")
ans
```

```python
# Sorting number of downloads for medals

ans = ps.sqldf("""

SELECT f.medal,
SUM(f.downloads)
FROM df AS f
GROUP BY f.medal
ORDER BY f.downloads DESC;

""")
ans
```

```python
# Getting the records containing name 'Cricket' or '3D' in title

ans = ps.sqldf("""

SELECT *
FROM df AS f
WHERE f.title LIKE '%Cricket%' OR f.title LIKE '%3D%';

""")
ans
```
